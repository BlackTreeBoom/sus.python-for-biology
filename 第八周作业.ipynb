```python
new1=gene_phenotype.sort_values('score',ascending=False)#先对表达的基因进行操作
```

```python
positive10=new1.head(10).copy()#取出前后面的十行，表达为positive,出现了报错，原因在于数据来源于视图
positive10= positive10.reset_index(drop = True)#为了消除视图的影响，使用copy
positive10['hit_class'] = 'positive'

negative10=new1.tail(10).copy()#取出前后面的十行，表达为positive,出现了报错，原因在于数据来源于视图
negative10= negative10.reset_index(drop = True)
negative10['hit_class'] = 'negative'
```

```python
workplace=pd.merge(positive10,sgRNA_score)#合并列表
```

```python
d=workplace2.sort_values('sgRNA_score',ascending=False)
#d=d.groupby('single').first()#先进行整体的sort_values，在groupy取值
```

```python
positive_gene = workplace2.groupby('Gene').head(3)#head（）取前面几个
```

```python
workplace=pd.merge(negative10,sgRNA_score)
workplace2=pd.merge(workplace,CRISPRi_v2)
d=workplace2.sort_values('sgRNA_score',ascending=False)
#d=d.groupby('single').first()#先进行整体的sort_values，在groupy取值
nagetive_gene = workplace2.groupby('Gene').head(3)
```

```python
gene10=pd.concat([positive_gene,nagetive_gene])
```
