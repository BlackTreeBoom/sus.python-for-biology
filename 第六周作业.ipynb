```python
class  Atom:#定义Atom的Class
    def __init__(self,Atom_name,Coordinates):#_init_函数用于初始化，self相当于整个Atom_Class自己，可以自用命名
        self.Atom_name = Atom_name#之所以用self为了后边便捷的调用里面的Atom_name,Coordinates元素
        self.Coordinates=Coordinates#正如老师上课所讲的方法，第一个是汽车名，后面是轮胎呀，汽油呀啥的汽车的参数

class Residue:
    def __init__(self,Residue_name,Residue_sequence_number,Atoms):
        self.Residue_name = Residue_name
        self.Residue_sequence_number=Residue_sequence_number
        self.Atoms=Atoms
    def List_all_Atoms(self):
        alist=[]#创建空列表
        for x in self.Atoms:#遍历列表中的每一个元素
            alist.append(x.Atom_name)
        return alist

class Chain:
    def __init__(self,Chain_name,Residues):
        self.Chain_name=Chain_name
        self.Residues=Residues
        print("aaaaaa")
    def Total_number_of_residues(self):
        number_count1 = self.Residues.count('Residues')
        print('The total number of residues is',number_count1)
    def List_all_residue_names(self):
        alist=[]#创建空列表
        for x in self.Residues:#遍历列表中的每一个元素
            alist.append(x.Residue_name)
        return alist
        
class Struture:
    def __init__(self,PDF_identifier,Chains):
        self.PDF_identifier=PDF_identifier
        self.Chains=Chains
    def Total_number_of_chains(self):
        number_count2 = len(self.Chains)
        print('The total number of chains is',number_count2)
    def Total_number_of_residues(self):
        sum=0
        for x in self.Chains:
            number_count3 = x.Total_number_of_residues()
            sum = sum + number_count3
        print('The total number of residues is',sum)
    def List_all_residues(self):
        clist=[]#创建空列表
        for x in self.Chains:#遍历列表中的每一个元素
            alist.append(x.List_all_residue_names)
```

```python
with open ('d://3j5p.pdb') as f:#通过withopen函数打开文件,两个斜杠\或者//
    listwork=f.readline()#第一行直接读
    listwork1=listwork[62:66]
    lines=f.readlines()#readlins是个函数，需要（）
s1=Struture(listwork1,Chains=[])#Struture(listwork1,Chains=[])是一个存储信息的地址，需要给它赋予简单的名字
s1.PDF_identifier
```

```python
newlist=[]#建立一个二重列表的空列表，往里面加列表
for i in range(1,len(lines)):
    line=lines[i-1].split(' ')#去掉\n符号
    if 'ATOM' in line:#用于判断STR中是否含有'ATOM'字符串
        tmplist = filter(None, line)#filter函数用于python去掉所有空格
        newline = list(tmplist)#处理后变为了二重列表
        newlist.append(newline)
del(newlist[0])
        #删掉第一个干扰元素

n=0
j=0
for i in range(1,len(newlist)-1):#考虑len长度，得从1开始 
    if i == 1:
        s1.Chains.append(Chain(newlist[i-1][4],Residues=[]))
        s1.Chains[0].Residues.append(Residue(newlist[i-1][3],newlist[i-1][1],Atoms=[]))
        s1.Chains[0].Residues[0].Atoms.append(Atom(newlist[i-1][2],Coordinates=newlist[i-1][6:9]))#append只能添加一个对象元素
    else:
        if newlist[i-1][4] != newlist[i][4]:#先判断最外层的,如果不相等，创建一个新的Chain加入Chains
            n=n+1
            j=0#重置Residues的位置，新的Chain必然是新的Residues
            s1.Chains.append(Chain(newlist[i][4],Residues=[]))#不相等，添加新的Chain，n+1下一次遍历要从N+1开始
            s1.Chains[n].Residues.append(Residue(newlist[i][3],newlist[i][1],Atoms=[]))
            s1.Chains[n].Residues[j].Atoms.append(Atom(newlist[i][2],Coordinates=newlist[i-1][6:9]))#append只能添加一个对象元素
        else:
            if newlist[i-1][3] != newlist[i][3]:
                j=j+1#这里创建新的Residue因此j要+1
                s1.Chains[n].Residues.append(Residue(newlist[i][3],newlist[i][1],Atoms=[]))
                s1.Chains[n].Residues[j].Atoms.append(Atom(newlist[i][2],Coordinates=newlist[i-1][6:9]))#append只能添加一个对象元素
            else:
                s1.Chains[n].Residues[j].Atoms.append(Atom(newlist[i][2],Coordinates=newlist[i-1][6:9]))#append只能添加一个对象元素
print(n)
```
